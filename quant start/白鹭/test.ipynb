```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "./Question4.csv"
df = pd.read_csv(filepath)
df
```

```python
df.loc[df["iid"] == 1]
```

```python
df["mid_price"] = (df["high"] + df["low"]) / 2
```

```python
def moving_avg(look_back_window, column):
    iids = df["iid"].unique()
    new_col = pd.DataFrame(np.zeros(len(df)), columns=["value"])
    for iid in iids:
        new_col.loc[df["iid"] == iid, "value"] = df.loc[df["iid"] == iid, column].rolling(window=look_back_window).mean()
    
    return new_col["value"]
```

```python
df["MA3_CLOSE"] = moving_avg(3, "close")
df["MA5_VOLUME"] = moving_avg(5, "volume")
```

```python
df
```

```python
def ret(time_diff):
    iids = df["iid"].unique()
    new_col = pd.DataFrame(np.zeros(len(df)), columns=["value"])
    for iid in iids:
        new_col.loc[df["iid"] == iid, "value"] = df.loc[df["iid"] == iid, "close"].shift(-time_diff) / df.loc[df["iid"] == iid, "close"] - 1
    return new_col["value"]
```

```python
df["Ret30"] = ret(30)
df["Ret240"] = ret(240)
```

```python
df
```

```python
features = ["MA3_CLOSE", "MA5_VOLUME", "Ret30", "Ret240"]
corr_matrix = df[features].corr(method='pearson')
print(corr_matrix)
'''
The Result is
            MA3_CLOSE  MA5_VOLUME     Ret30    Ret240
MA3_CLOSE    1.000000   -0.307805 -0.031818 -0.091880
MA5_VOLUME  -0.307805    1.000000  0.011852 -0.004616
Ret30       -0.031818    0.011852  1.000000  0.350430
Ret240      -0.091880   -0.004616  0.350430  1.000000
'''
```
