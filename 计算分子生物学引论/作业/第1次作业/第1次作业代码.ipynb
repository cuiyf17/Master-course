```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

# 计算外显子长度分布

```python
def exon_len_distribution(exon_file = "genome_exon.txt"):
    chr_list = ['chr1', 'chr2', 'chr3', 'chr4', 'chr5', 'chr6', 'chr7', 'chr8', 'chr9', 'chr10', 'chr11', 'chr12', 'chr13', 'chr14', 'chr15', 'chr16', 'chr17', 'chr18', 'chr19', 'chr20', 'chr21', 'chr22', 'chrX', 'chrY']
    genome_length_record = []
    chr_dict = dict()
    area = None
    length = 0
    for count, line in enumerate(open("./length_distribution/" + exon_file, 'r')):
        if(line.startswith(">")):
            if(area is not None):
                if(area not in chr_dict):
                    chr_dict[area] = list()
                chr_dict[area].append(length)
                genome_length_record.append(length)
            length = 0
            for chr_ in chr_list:
                if(chr_+":" in line):
                    area = chr_
                    break
        else:
            length+=len(line)-1

    for area in chr_dict:
        chr_dict[area] = np.array(chr_dict[area])
        print(area+"的外显子长度分布为：")
        plt.figure(figsize=(10,5))
        plt.title(area+" exon length distribution (min = %d, max = %d, median = %d, mean = %.2f)"%(np.min(chr_dict[area]), np.max(chr_dict[area]), np.median(chr_dict[area]), np.mean(chr_dict[area])))
        plt.hist(chr_dict[area], bins=1000, facecolor="blue", edgecolor="black")
        plt.xscale('log')
        plt.xlabel("length")
        plt.ylabel("count")
        plt.show()
    
    area = "The whole genome"
    genome_length_record = np.array(genome_length_record)
    print(area+"的外显子长度分布为：")
    plt.figure(figsize=(10,5))
    plt.title(area+" exon length distribution (min = %d, max = %d, median = %d, mean = %.2f)"%(genome_length_record.min(), genome_length_record.max(), np.median(genome_length_record), genome_length_record.mean()))
    plt.hist(genome_length_record, bins=5000, facecolor="blue", edgecolor="black")
    plt.xscale('log')
    plt.xlabel("length")
    plt.ylabel("count")
    plt.show()
    
exon_len_distribution()
```
