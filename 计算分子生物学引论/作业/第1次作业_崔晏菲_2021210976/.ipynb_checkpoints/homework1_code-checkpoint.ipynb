```python
from gtfparse import read_gtf
import matplotlib.pylab as plt
import numpy as np
import pandas as pd
```

### 1. 读取数据

```python
data = read_gtf("./gencode.v41.annotation.gtf.gz")
```

```python
print(data.iloc[0])
```

### 2. 计算染色体上的蛋白质编码基因

```python
# 生成染色体编号
chr_num = ["chr%d"%(i) for i in range(1,23)]
chr_num.extend(["chrX", "chrY", "chrM"])

protein_coding_genes = dict()
for chr in chr_num:
    if(chr not in protein_coding_genes):
        protein_coding_genes[chr] = -1

# 计算每个染色体上的蛋白质编码基因
for chr in chr_num:
    chr_data = data[data["seqname"] == chr]
    chr_gene = chr_data[chr_data["feature"] == "gene"]
    chr_protein_gene = chr_gene[chr_gene["gene_type"] == "protein_coding"]
    protein_coding_genes[chr] = len(chr_protein_gene)



print(protein_coding_genes)
```

```python
def cal_len(gene, ind):
    len = gene["end"][ind] - gene["start"][ind] + 1
    return len

def cal_dis(gene_data, chr_data):
    genes_len_list, exons_len_list, introns_len_list = [], [], []
    for i in gene_data.index:
        gene_len = cal_len(chr_data, i)
        exon_len = 0
        if chr_data["feature"][i + 1] == "transcript":
            exon_end = chr_data["start"][i + 1]
            trains_len = cal_len(chr_data, i + 1)
            i += 2
            while chr_data["feature"][i] != "gene" and chr_data["feature"][i] != "transcript":
                if chr_data["feature"][i] == "exon":
                    if chr_data["start"][i] >= exon_end:
                        exon_len += cal_len(chr_data, i)
                        exon_end = chr_data["end"][i]
                    else:
                        break
                i += 1
            genes_len_list.append(gene_len)
            exons_len_list.append(exon_len)
            introns_len_list.append(gene_len - exon_len)
    return genes_len_list, exons_len_list, introns_len_list           
```

```python
# 计算相关的长度数据
chr_genes_len = []
chr_exons_len = []
chr_introns_len = []
for chr in chr_num:
    chr_data = data[data["seqname"] == chr]
    chr_genes = chr_data[chr_data["feature"] == "gene"]
    chr_protein_genes = chr_genes[chr_genes["gene_type"] == "protein_coding"]
    genes_len_list, exons_len_list, introns_len_list = cal_dis(chr_protein_genes, chr_data)
    chr_genes_len.append(genes_len_list)
    chr_exons_len.append(exons_len_list)
    chr_introns_len.append(introns_len_list)
```

```python
# 读取全部染色体的长度数据
whole_genes_len = sum(chr_genes_len, [])
whole_exons_len = sum(chr_exons_len, [])
whole_introns_len = sum(chr_introns_len, [])

whole_genes_len = np.array(whole_genes_len)
whole_exons_len = np.array(whole_exons_len)
whole_introns_len = np.array(whole_introns_len)

whole_genes_len_log = np.log(whole_genes_len)
whole_exons_len_log = np.log(whole_exons_len)
whole_introns_len_log = np.log(whole_introns_len)

print("median, mean, max, min")
print("the statistics of genes: ", np.median(whole_genes_len), np.mean(whole_genes_len), np.max(whole_genes_len), np.min(whole_genes_len))
print("the statistics of exons: ", np.median(whole_exons_len), np.mean(whole_exons_len), np.max(whole_exons_len), np.min(whole_exons_len))
print("the statistics of introns: ", np.median(whole_introns_len), np.mean(whole_introns_len), np.max(whole_introns_len), np.min(whole_introns_len))

plt.figure(0, figsize=(12, 18))
plt.subplot(3,2,1)
plt.title("the length of genes")
plt.xlabel("length/bp")
plt.ylabel("nums")
plt.hist(whole_genes_len, 100, range=(0, whole_genes_len.max()))
plt.subplot(3,2,3)
plt.title("the length of exons")
plt.xlabel("length/bp")
plt.ylabel("nums")
plt.hist(whole_exons_len, 100, range=(0, whole_exons_len.max()))
plt.subplot(3,2,5)
if whole_introns_len.max() != 0:
    plt.title("the length of introns")
    plt.xlabel("length/bp")
    plt.ylabel("nums")
    plt.hist(whole_introns_len, 100, range=(0, whole_introns_len.max()))

plt.subplot(3,2,2)
plt.title("the length of genes")
plt.xlabel("log(length)/bp")
plt.ylabel("nums")
plt.hist(whole_genes_len_log, 100, range=(0, whole_genes_len_log.max()))
plt.subplot(3,2,4)
plt.title("the length of exons")
plt.xlabel("log(length)/bp")
plt.ylabel("nums")
plt.hist(whole_exons_len_log, 100, range=(0, whole_exons_len_log.max()))
plt.subplot(3,2,6)
if whole_introns_len.max() != 0:
    plt.title("the length of introns")
    plt.xlabel("log(length)/bp")
    plt.ylabel("nums")
    plt.hist(whole_introns_len_log, 100, range=(0, whole_introns_len_log.max()))
plt.savefig('whole_gene.png')
```

```python
from cmath import inf
# median mean max min
gene_stat = [[],[],[],[]] 
exon_stat = [[],[],[],[]]
intron_stat = [[],[],[],[]]


for i in range(len(chr_num)):
    # 依次读取每个染色体的数据
    genes_len = np.array(chr_genes_len[i])
    exons_len = np.array(chr_exons_len[i])
    introns_len = np.array(chr_introns_len[i])
    
    # 计算每个染色特的统计数据
    gene_stat[0].append(np.median(genes_len))
    gene_stat[1].append(np.mean(genes_len))
    gene_stat[2].append(np.max(genes_len))
    gene_stat[3].append(np.min(genes_len))
    exon_stat[0].append(np.median(exons_len))
    exon_stat[1].append(np.mean(exons_len))
    exon_stat[2].append(np.max(exons_len))
    exon_stat[3].append(np.min(exons_len))
    intron_stat[0].append(np.median(introns_len))
    intron_stat[1].append(np.mean(introns_len))
    intron_stat[2].append(np.max(introns_len))
    intron_stat[3].append(np.min(introns_len))
    
    # 对长度取对数方便展示
    genes_len_log = np.log10(np.array(chr_genes_len[i]))
    exons_len_log = np.log10(np.array(chr_exons_len[i]))
    introns_len_log = np.log10(np.array(chr_introns_len[i]))
    # 画出长度的分布直方图
    plt.figure(i + 1, figsize=(12, 16))
    plt.subplot(3,2,1)
    plt.title("the length of genes in " + chr_num[i])
    plt.xlabel("length/bp")
    plt.ylabel("nums")
    plt.hist(genes_len, 100, range=(0, genes_len.max()))
    plt.subplot(3,2,3)
    plt.title("the length of exons in " + chr_num[i])
    plt.xlabel("length/bp")
    plt.ylabel("nums")
    plt.hist(exons_len, 100, range=(0, exons_len.max()))
    plt.subplot(3,2,5)
    if introns_len.max() != 0:
        plt.title("the length of introns in " + chr_num[i])
        plt.xlabel("length/bp")
        plt.ylabel("nums")
        plt.hist(introns_len, 100, range=(0, introns_len.max()))
    
    plt.subplot(3,2,2)
    plt.title("the length of genes in " + chr_num[i])
    plt.xlabel("log(length)/bp")
    plt.ylabel("nums")
    plt.hist(genes_len_log, 100, range=(0, genes_len_log.max()))
    plt.subplot(3,2,4)
    plt.title("the length of exons in " + chr_num[i])
    plt.xlabel("log(length)/bp")
    plt.ylabel("nums")
    plt.hist(exons_len_log, 100, range=(0, exons_len_log.max()))
    plt.subplot(3,2,6)
    if introns_len.max() != 0:
        plt.title("the length of introns in " + chr_num[i])
        plt.xlabel("log(length)/bp")
        plt.ylabel("nums")
        plt.hist(introns_len_log, 100, range=(0, introns_len_log.max()))
    plt.savefig("hw1_fig/"+chr_num[i]+".png")

```

```python
np.savetxt("gene_stat.csv", gene_stat, fmt="%f", delimiter=",")
np.savetxt("exon_stat.csv", exon_stat, fmt="%f", delimiter=",")
np.savetxt("intron_stat.csv", intron_stat, fmt="%f", delimiter=",")
```
