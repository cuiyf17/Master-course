```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

# 计算外显子长度分布

```python
#file_list = [file for file in os.listdir("./length_distribution/exons/") if file.endswith(".txt")]
file_list = ['chr1.txt', 'chr2.txt', 'chr3.txt', 'chr4.txt', 'chr5.txt', 'chr6.txt', 'chr7.txt', 'chr8.txt', 'chr9.txt', 'chr10.txt', 'chr11.txt', 'chr12.txt', 'chr13.txt', 'chr14.txt', 'chr15.txt', 'chr16.txt', 'chr17.txt', 'chr18.txt', 'chr19.txt', 'chr20.txt', 'chr21.txt', 'chr22.txt', 'chrX.txt', 'chrY.txt']
genome_length_record = []
for file in file_list:
    area = file.replace(".txt", "")
    length_record = []
    with open("./length_distribution/exons/"+file) as f:
        line = f.readline()
        while(line):
            length = 0
            if(line.startswith(">")):
                line = f.readline()
            else:
                length+=len(line)-1
                line = f.readline()
                while(line):
                    if(line.startswith(">")):
                        break
                    else:
                        length+=len(line)-1
                        line = f.readline()
                length_record.append(length)
                genome_length_record.append(length)
    length_record = np.array(length_record)
    print(area+"的外显子长度分布为：")
    plt.figure(figsize=(10,5))
    plt.title(area+" exon length distribution (min = %d, max = %d, median = %d, mean = %.2f)"%(length_record.min(), length_record.max(), np.median(length_record), length_record.mean()))
    plt.hist(length_record, bins=400, facecolor="blue", edgecolor="black")
    plt.xscale('log')
    plt.xlabel("length")
    plt.ylabel("count")
    plt.show()
    
area = "The whole genome"
genome_length_record = np.array(genome_length_record)
print(area+"的外显子长度分布为：")
plt.figure(figsize=(10,5))
plt.title(area+" exon length distribution (min = %d, max = %d, median = %d, mean = %.2f)"%(genome_length_record.min(), genome_length_record.max(), np.median(genome_length_record), genome_length_record.mean()))
plt.hist(genome_length_record, bins=4000, facecolor="blue", edgecolor="black")
plt.xscale('log')
plt.xlabel("length")
plt.ylabel("count")
plt.show()
```
