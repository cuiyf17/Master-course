```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

# 1. 读取数据

```python
def load_ICU_data(data_path, label_path):
    if(not data_path.endswith(".csv")):
        print("数据需要csv格式的文件！")
        return None
    if(not label_path.endswith(".csv")):
        print("标签需要csv格式的文件！")
        return None
    data_df = pd.read_csv(data_path)
    label_df = pd.read_csv(label_path)
    data = np.array(data_df)
    label = np.array(label_df)
    return data, label
```

# 2. KNN

```python
train1_data, train1_label = load_ICU_data("./data1forEx1to4/train1_icu_data.csv", "./data1forEx1to4/train1_icu_label.csv")
test1_data, test1_label = load_ICU_data("./data1forEx1to4/test1_icu_data.csv", "./data1forEx1to4/test1_icu_label.csv")
train2_data, train2_label = load_ICU_data("./data1forEx1to4/train2_icu_data.csv", "./data1forEx1to4/train2_icu_label.csv")
test2_data, test2_label = load_ICU_data("./data1forEx1to4/test2_icu_data.csv", "./data1forEx1to4/test2_icu_label.csv")
train_data = np.concatenate([train1_data, train2_data], axis = 0)
train_label = np.concatenate([train1_label, train2_label], axis = 0)
test_data = np.concatenate([test1_data, test2_data], axis = 0)
test_label = np.concatenate([test1_label, test2_label], axis = 0)
```

```python
def K_fold_divide(X, Y, K=5):
    classes = np.unique(Y).reshape(-1)
    Y=Y.reshape((-1,1))
    folds = []
    for item in classes:
        x = X[np.where(Y[:,0]==item)]
        y = Y[np.where(Y[:,0]==item)]
        data = np.concatenate([x,y.reshape((-1,1))], axis=1)
        np.random.shuffle(data)
        M=data.shape[0]//K
        if((data.shape[0]-M*K)>0):
            M+=1
    
        for i in range(K):
            valid = data[i*M:min((i+1)*M, data.shape[0])]
            train = np.concatenate([data[0:M*i],data[min((i+1)*M, data.shape[0]):data.shape[0]]], axis = 0)
            train_data = train[:,:data.shape[1]-1]
            train_label = train[:,data.shape[1]-1]
            valid_data = valid[:,:data.shape[1]-1]
            valid_label = valid[:,data.shape[1]-1]
            if(len(folds)<K):
                folds.append([train_data, train_label, valid_data, valid_label])
            else:
                folds[i][0] = np.concatenate([folds[i][0], train_data], axis=0)
                folds[i][1] = np.concatenate([folds[i][1], train_label], axis=0)
                folds[i][2] = np.concatenate([folds[i][2], valid_data], axis=0)
                folds[i][3] = np.concatenate([folds[i][3], valid_label], axis=0)
    
    return folds
```

```python
# 进行K折交叉验证，并对数据做标准化预处理
n_folds = 3
mean = np.mean(train1_data,axis=0,keepdims=True)
std = np.std(train1_data,axis=0,keepdims=True)
train1_X = (train1_data - mean)/(std + 1e-12) 
test1_X = (test1_data - mean)/(std + 1e-12) 
folds = K_fold_divide(train1_X, train1_label, K = n_folds)
```

选择参数为weight = 'distance':

```python
print("%-17s%-17s%-17s%-17s%-17s%-17s%-17s%-17s%-17s%-17s"%("n_neighbors", "train accuracy", "train sensi", "train speci", "valid accuracy", "valid sensi", "valid speci", "test accuracy", "test sensi", "test speci"))
train_accus = []
valid_accus = []
test_accus = []
neighbors = range(1, 60)
for n_neighbors in neighbors:
    knns = []
    train_accu = []
    train_sensi = []
    train_speci = []
    valid_accu = []
    valid_sensi = []
    valid_speci = []
    test_accu = []
    test_sensi = []
    test_speci = []
    
    for i in range(n_folds):
        train_data, train_label, valid_data, valid_label = folds[i]
        knns.append(KNeighborsClassifier(n_neighbors=n_neighbors, weights = "distance"))
        knns[i].fit(train_data, train_label.reshape(-1))
        tn, fp, fn, tp = confusion_matrix(train_label.reshape(-1), knns[i].predict(train_data)).ravel()
        train_accu.append((tn+tp)/(tn+fp+fn+tp))
        train_sensi.append(tp/(tp+fn))
        train_speci.append(tn/(tn+fp))
    
        tn, fp, fn, tp = confusion_matrix(valid_label.reshape(-1), knns[i].predict(valid_data)).ravel()
        valid_accu.append((tn+tp)/(tn+fp+fn+tp))
        valid_sensi.append(tp/(tp+fn))
        valid_speci.append(tn/(tn+fp))
        
        tn, fp, fn, tp = confusion_matrix(test1_label.reshape(-1), knns[i].predict(test1_X)).ravel()
        test_accu.append((tn+tp)/(tn+fp+fn+tp))
        test_sensi.append(tp/(tp+fn))
        test_speci.append(tn/(tn+fp))
        #print("%-8d%-17f%-17f%-17f%-17f%-17f%-17f"%(i+1, train_accu[i], train_sensi[i], train_speci[i], valid_accu[i], valid_sensi[i], valid_speci[i]))
    mean_train_accu = np.mean(train_accu)
    mean_valid_accu = np.mean(valid_accu)
    mean_test_accu = np.mean(test_accu)
    train_accus.append(mean_train_accu)
    valid_accus.append(mean_valid_accu)
    test_accus.append(mean_test_accu)
    print("%-17d%-17f%-17f%-17f%-17f%-17f%-17f%-17f%-17f%-17f"%(n_neighbors, mean_train_accu, np.mean(train_sensi), np.mean(train_speci), mean_valid_accu, np.mean(valid_sensi), np.mean(valid_speci), mean_test_accu, np.mean(test_sensi), np.mean(test_speci)))
```

# 3. MLP

```python
def calculate_ROC(output, label):
    Y = np.where(label==0, 1, -1).reshape(-1,1)
    thresholds = np.concatenate([np.zeros(1), np.unique(output), np.ones(1)])
    TPRs = []
    FPRs = []
    AUC = 0
    for i in range(thresholds.shape[0]):
        threshold = thresholds[i]
        Y_predicted = np.where(output <= threshold, 1, -1)
        Y_predicted = np.where(Y_predicted > 0, 2, -1)
        TPR = (Y_predicted == Y).sum()/((Y==-1).sum())
        Y_predicted = np.where(Y_predicted > 0, 1, -2)
        FPR = 1-(Y_predicted == Y).sum()/(Y==1).sum()
        TPRs.append(TPR)
        FPRs.append(FPR)
        if(threshold>0):
            AUC+=(TPRs[i]+TPRs[i-1])*(-FPRs[i]+FPRs[i-1])/2
            
    return TPRs, FPRs, AUC
```

```python
class SoftmaxCrossEntropyLossLayer():
    def __init__(self,num_classes):
        self.acc = 0.
        self.loss = 0.
        self.num_classes=num_classes

    def forward(self, X, Y):    
        self.Y=Y
        self.Y_one_hot = (np.arange(self.num_classes)==Y)
        X_max=np.max(X,axis=1,keepdims=True)     
        X=np.exp(X-X_max)
        self.probability=(X/np.sum(X,axis=1,keepdims=True))
        prediction=np.argmax(self.probability,axis=1).reshape(-1,1)
        error=-np.sum(self.Y_one_hot*np.log(self.probability+EPS),axis=1,keepdims=True)
        self.loss=np.mean(error)
        self.acc=np.mean(prediction == Y)
        return self.loss

    def backward(self):
        return self.probability-self.Y_one_hot
```

```python
class FCLayer():
    def __init__(self, num_input, num_output, actFunction='relu', trainable=True):
        self.num_input = num_input
        self.num_output = num_output
        self.trainable = trainable
        self.actFunction = actFunction
        self.V_dw=0
        self.V_db=0
        assert actFunction in ['relu', 'sigmoid']
        self.XavierInit()

    def forward(self, Input):
        self.input = Input
        output = np.dot(self.input, self.W) + self.b      
        return output

    def backward(self, delta):
        
        gradient_delta=(np.dot(self.W,delta.T)).T
        self.delta=gradient_delta

        N = delta.shape[0]
        self.grad_W = np.dot(self.input.T, delta) / N
        self.grad_b = np.sum(delta, axis=0, keepdims=True) / N
        return gradient_delta

    def XavierInit(self):
        raw_std = (2 / (self.num_input + self.num_output))**0.5
        if 'relu' == self.actFunction:
            init_std = raw_std * (2**0.5)
        elif 'sigmoid' == self.actFunction:
            init_std = raw_std
        else:
            init_std = raw_std # * 4

        self.W = np.random.normal(0, init_std, (self.num_input, self.num_output))
        self.b = np.random.normal(0, init_std, (1, self.num_output))
```

```python
class ReLULayer():
    def __init__(self):
        self.trainable = False # no parameters
        self.V_dw=0
        self.V_db=0
        
    def forward(self, Input):
        self.input = Input    
        return ((Input+abs(Input))/2.0)

    def backward(self, delta):
        self.delta=delta*np.maximum(0,self.input)   
        return self.delta
```

```python
class SigmoidLayer():
    def __init__(self):
        self.trainable = False
        self.V_dw=0
        self.V_db=0
        
    def forward(self, Input):
        self.output = 1 / (1 + np.exp(-Input))   
        return self.output

    def backward(self, delta):        
        return self.output * (1 - self.output) * delta
```

```python
class Network():
    def __init__(self):
        self.layerList = []
        self.numLayer = 0
        self.train_accuracy_list=[]
        self.test_accuracy_list=[]
        self.train_loss_list=[]
        self.test_loss_list=[]
        self.train_index_list=[]
        self.test_index_list=[]
        self.train_FPR=[]
        self.train_TPR=[]
        self.test_FPR=[]
        self.test_TPR=[]
        self.train_AUC=0
        self.test_AUC=0
    def add(self, layer):
        self.numLayer += 1
        self.layerList.append(layer)

    def forward(self, x):
        for i in range(self.numLayer):
            x = self.layerList[i].forward(x)
        return x

    def backward(self, delta):
        for i in reversed(range(self.numLayer)):
            delta = self.layerList[i].backward(delta)
```

```python
class MiniBatchwithMomentum():
    def __init__(self, learningRate, weightDecay, momentum=0.9):
        self.learningRate = learningRate
        self.weightDecay = weightDecay
        self.momentum = momentum

    def step(self, model):
        layers = model.layerList
        for layer in layers:
            if layer.trainable:
                layer.diff_W = (layer.grad_W + self.weightDecay * layer.W)
                layer.diff_b = layer.grad_b
                layer.V_dw = self.momentum*layer.V_dw+(1-self.momentum)*layer.diff_W
                layer.V_db = self.momentum*layer.V_db+(1-self.momentum)*layer.diff_b
                layer.W -= self.learningRate*layer.V_dw
                layer.b -= self.learningRate*layer.V_db
```

```python
def TrainModel(num_classes,learning_rate,weight_decay,max_epoch,batch_size,model,criterion,optimizer):
    num_batch=train1_X.shape[0]//batch_size
    if(train1_X.shape[0]!=batch_size*num_batch):
        num_batch+=1
    for epoch in range(max_epoch):
        print("============================== epoch %d =============================="%(epoch))
        for i in range(num_batch):
            X=model.forward(train1_X[i*batch_size:(i+1)*batch_size])
            Y=train1_label[i*batch_size:(i+1)*batch_size].reshape(-1,1)
            criterion.forward(X, Y)
            delta = criterion.backward()
            model.backward(delta)
            optimizer.step(model)
            
            #model.train_accuracy_list.append(criterion.acc)
            #model.train_loss_list.append(criterion.loss)
            if(num_classes==2):
                Y_predicted=np.argmax(criterion.probability,axis=1).reshape(-1,1)       
                Y_tmp=np.where(Y < 0.5, -1, 1) 
                TPR=(Y_predicted == Y_tmp).sum()/(Y_tmp==1).sum()    
                Y_tmp=np.where(Y > 0.5, 2, 0)
                TNR=(Y_predicted == Y_tmp).sum()/(Y_tmp==0).sum()       
                FPR=1-TNR
                FNR=1-TPR     
                specificity=TNR
                sensitivity=TPR
                Y_predicted_tmp=np.where(Y_predicted > 0.5, 1, -1)
                FDR=1-(Y_predicted_tmp == Y).sum()/(Y_predicted_tmp==1).sum()
                #model.train_index_list.append([sensitivity, specificity, TPR, TNR, FPR, FNR, FDR])
                print("batch %d/%d\taccu = %.4f, loss = %.4f, sensitivity = %.4f, specificity = %.4f, TPR = %.4f, TNR = %.4f, FPR = %.4f, FNR = %.4f, FDR = %.4f"%(i*batch_size, train1_data.shape[0], criterion.acc, criterion.loss, sensitivity, specificity, TPR, TNR, FPR, FNR, FDR))
            else:
                print("batch %d/%d\taccu = %.4f, loss = %.4f"%(i*batch_size, train1_data.shape[0], criterion.acc, criterion.loss))
                
        X=model.forward(train1_X)
        Y=train1_label.reshape(-1,1)
        criterion.forward(X, Y)
        print("------------------------- Average Performance -------------------------")
        model.train_accuracy_list.append(criterion.acc)
        model.train_loss_list.append(criterion.loss)
        if(num_classes==2):
            Y_predicted=np.argmax(criterion.probability,axis=1).reshape(-1,1)       
            Y_tmp=np.where(Y < 0.5, -1, 1) 
            TPR=(Y_predicted == Y_tmp).sum()/(Y_tmp==1).sum()    
            Y_tmp=np.where(Y > 0.5, 2, 0)
            TNR=(Y_predicted == Y_tmp).sum()/(Y_tmp==0).sum()       
            FPR=1-TNR
            FNR=1-TPR     
            specificity=TNR
            sensitivity=TPR
            Y_predicted_tmp=np.where(Y_predicted > 0.5, 1, -1)
            FDR=1-(Y_predicted_tmp == Y).sum()/(Y_predicted_tmp==1).sum()
            print("accu = %.4f, loss = %.4f, sensitivity = %.4f, specificity = %.4f, TPR = %.4f, TNR = %.4f, FPR = %.4f, FNR = %.4f, FDR = %.4f"%(criterion.acc, criterion.loss, sensitivity, specificity, TPR, TNR, FPR, FNR, FDR))
        else:
            print("accu = %.4f, loss = %.4f"%(criterion.acc, criterion.loss))
           
    if(num_classes==2):
        X=model.forward(train1_X)
        Y=train1_label.reshape(-1,1)
        criterion.forward(X, Y)
        Y_predicted=np.array(criterion.probability[:,1].reshape(-1,1))
        TPRs, FPRs, AUC = calculate_ROC(Y_predicted, Y)
        model.train_FPR = FPRs
        model.train_TPR = TPRs
        model.train_AUC=AUC
        print("train_AUC = %.4f"%(AUC))
```

```python
def TestModel(num_classes,model,criterion):
    X=model.forward(test1_X)
    Y=test1_label.reshape(-1,1)
    criterion.forward(X, Y)
    model.test_accuracy_list.append(criterion.acc)
    model.test_loss_list.append(criterion.loss)
    if(num_classes==2):
        Y_predicted=np.argmax(criterion.probability,axis=1).reshape(-1,1)       
        Y_tmp=np.where(Y < 0.5, -1, 1) 
        TPR=(Y_predicted == Y_tmp).sum()/(Y_tmp==1).sum()    
        Y_tmp=np.where(Y > 0.5, 2, 0)
        TNR=(Y_predicted == Y_tmp).sum()/(Y_tmp==0).sum()       
        FPR=1-TNR
        FNR=1-TPR     
        specificity=TNR
        sensitivity=TPR
        Y_predicted_tmp=np.where(Y_predicted > 0.5, 1, -1)
        FDR=1-(Y_predicted_tmp == Y).sum()/(Y_predicted_tmp==1).sum()
        model.test_index_list.append([sensitivity, specificity, TPR, TNR, FPR, FNR, FDR])
        print("accu = %.4f, loss = %.4f, sensitivity = %.4f, specificity = %.4f, TPR = %.4f, TNR = %.4f, FPR = %.4f, FNR = %.4f, FDR = %.4f"%(criterion.acc, criterion.loss, sensitivity, specificity, TPR, TNR, FPR, FNR, FDR))
    else:
        print("accu = %.4f, loss = %.4f"%(criterion.acc, criterion.loss))
    if(num_classes==2):
        X=model.forward(test1_X)
        Y=test1_label.reshape(-1,1)
        criterion.forward(X, Y)
        Y_predicted=np.array(criterion.probability[:,1].reshape(-1,1))
        TPRs, FPRs, AUC = calculate_ROC(Y_predicted, Y)
        model.test_FPR = FPRs
        model.test_TPR = TPRs
        model.test_AUC=AUC
        print("test_AUC = %.4f"%(AUC))
```

```python
def PlotModel(model, num_classes):
    y1 = model.train_accuracy_list
    y2 = model.train_loss_list
    y3 = model.train_accuracy_list
    y4 = model.train_accuracy_list
    
    plt.plot(range(len(model.train_accuracy_list)), model.train_accuracy_list, '-')
    plt.title('Train accuracy')
    plt.xlabel('iteration')
    plt.ylabel('accuracy')
    plt.show()
    
    plt.plot(range(len(model.train_loss_list)), model.train_loss_list, '-')
    plt.title('Train loss')
    plt.xlabel('iteration')
    plt.ylabel('loss')
    plt.show()
    
    if(num_classes==2):
        y1 = model.train_TPR
        y2 = model.test_TPR
        x1 = model.train_FPR
        x2 = model.test_FPR

        plt.plot(x1, y1, '-')
        plt.title('Train ROC')
        plt.xlabel('FPR')
        plt.ylabel('TPR')
        plt.show()

        plt.plot(x2, y2, '-')
        plt.title('Test ROC')
        plt.xlabel('FPR')
        plt.ylabel('TPR')
        plt.show()
```

```python
EPS = 1e-12
num_classes=2

max_epoch=70
learning_rate=0.1
weight_decay=0.0
batch_size=128


MLP=Network()
MLP.add(FCLayer(108,32))
MLP.add(ReLULayer())
MLP.add(FCLayer(32,num_classes))
criterion=SoftmaxCrossEntropyLossLayer(num_classes)
optimizer=MiniBatchwithMomentum(learning_rate,weight_decay)
```

```python
TrainModel(num_classes,learning_rate,weight_decay,max_epoch,batch_size,MLP,criterion,optimizer)
```
